1. Encode categorical variables to numerical.

```python
from sklearn.preprocessing import LabelEncoder

# List of categorical columns
categorical_cols = ['gender', 'respiratory_condition', 'fever_muscle_pain', 'status']

# Encode categorical columns
label_encoder = LabelEncoder()
for col in categorical_cols:
    df[col] = label_encoder.fit_transform(df[col])
```

```python
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
print("Numerical Columns:", numerical_cols)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col])
    plt.title(f"Box Plot for {col}")
    plt.show()
```

```python
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Example for 'duration'
outliers_duration = detect_outliers_iqr(df, 'duration')
print(f"Outliers in 'duration': {len(outliers_duration)}")
```

```python
from scipy import stats

def detect_outliers_zscore(data, column, threshold=3):
    z_scores = stats.zscore(data[column])
    outliers = data[(abs(z_scores) > threshold)]
    return outliers

# Example for 'rms_energy'
outliers_rms = detect_outliers_zscore(df, 'rms_energy')
print(f"Outliers in 'rms_energy': {len(outliers_rms)}")
```

```python
# Iterate through all numerical columns
for col in numerical_cols:
    # Skip the target variable if needed
    if col == 'cough_detected':
        continue
    
    # Detect outliers using IQR
    outliers_iqr = detect_outliers_iqr(df, col)
    print(f"Column: {col}")
    print(f"  Outliers (IQR method): {len(outliers_iqr)}")
    
    # Detect outliers using Z-score
    outliers_zscore = detect_outliers_zscore(df, col)
    print(f"  Outliers (Z-score method): {len(outliers_zscore)}")
    
    # Plot box plot
    plt.figure()
    sns.boxplot(x=df[col])
    plt.title(f"Box Plot for {col}")
    plt.show()
```
